# 分析数据的基本方法
## 分析流程
1. 引入所需的库
2. 读取 csv 文件
3. 认识数据, 认识数据, 认识数据(重要的事情说三遍). 包括其数据本身特征和牵扯的业务/场景
    - 一般需要了解如下指标/特征
        - 数据各指标的意思, 数据各值的意思.
        - 数据完整性
        - 数据大概的分布
    - 有时需要绘图查看数据的分布, 进一步了解数据
4. 数据分析
    - 预处理. 包括但不限于 清洗数据/填充空值.
    - 提出猜测
    - 验证猜测
5. 建立简单的模型, 分析验证. 除非得到合适的模型, 否则回归第三步.

## 数据处理的常用方法
### 特征缩放
特征缩放是用来统一资料中的自变项或特征范围的方法. 数据标准化就是特征缩放的一种.

特征缩放将数据集中各特征的缩放尺度标准化, 如特征A缩放尺度为千(即变化幅度多以千为单位), 特征B缩放尺度为十,
标准化后, 缩放范围都为 `[-1,1]`.

特征缩放后, 多维特征将具有相近的相似度, 对于 梯度下降/线性回归 等算法, 有助于提升收敛速度.

具体参考 [特征缩放](./特征缩放.ipynb)

### 差值填充
1. 差值较少可考虑手动填充
2. 平均值填充
3. 就近补齐: 选择数据相似的标准, 用其值填充缺值.
    - 如选择相邻最近的一条数据

## 建模建议
1. 了解数据, 了解数据的业务, 了解数据的特征.
2. 不必过早优化, 先做一个简单的模型, 然后逐步完善模型.
3. 注意数据的离散点
